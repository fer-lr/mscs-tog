```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python

nrm_df = pd.read_csv('SNR_Methods.csv', index_col = 0)
print(f"Total Method Count: {nrm_df.shape[0]}")
nrm_df = nrm_df.set_index('NAME')
```

```python
mtype_df = nrm_df[['TYPE']].TYPE.value_counts(dropna=True)

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#ffdd99','#33ddff']
ax = mtype_df.plot(kind="pie", autopct='%1.1f%%', colors=colors, legend=True, labeldistance=None)
ax.legend(bbox_to_anchor=(1, 1.02), loc='upper left')
display(nrm_df[['TYPE']].TYPE.value_counts(dropna=True))
plt.show()
```

```python
a_options = ['Filter',]
a = nrm_df[['TYPE','PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']]
a = a[a["TYPE"].isin(a_options)]
a = a[['PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']].mean(axis=0).to_frame(name="FILTERS")
a = a.dropna()

a

b_options = ['Deep Learning', 'CNN']
b = nrm_df[['TYPE','PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']]
b = b[b["TYPE"].isin(b_options)]
b = b[['PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']].mean(axis=0).to_frame(name="MACHINE")
b = b.dropna()


c_options = ['Wavelet']
c= nrm_df[['TYPE','PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']]
c = c[c["TYPE"].isin(c_options)]
c = c[['PSNR-10','PSNR-20','PSNR-25','PSNR-30','PSNR-40','PSNR-50','PSNR-60','PSNR-70','PSNR-80','PSNR-90']].mean(axis=0).to_frame(name="WAVELET")


d = a.join(b).join(c)

display(d)
sb.set_theme(style="whitegrid")
ax = sb.lineplot(d,markers=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 50)

plt.show()
```
